```python
from yelpapi import YelpAPI
import pandas as pd
import json
import os, json, math, time
with open('/Users/dmusl/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# set our API call parameters and filename before the first call
location = 'North Olmsted, OH 44070'
term = 'plant-based'
```

```python
location.split(',')[0]
```

```python
## Specify folder for saving data
FOLDER = 'Data/'

os.makedirs(FOLDER, exist_ok = True)
# Specifying JSON_FILE filename (can include a folder)
JSON_FILE = FOLDER+f"{location.split(',')[0]}-{term}.json"
```

```python
# use our yelp_api variable's search_query method to perform our API call
results = yelp_api.search_query(term = term, location = location)
```

```python
## how many businesses in our results
len(results['businesses'])
```

```python
pb_f = pd.DataFrame(results['businesses'])
pb_f.head() 
```

```python
# use our yelp_api variable's search_query method to perform our API call
results = yelp_api.search_query(term = term, location = location, offset = 20)
```

```python
pb_f2 = pd.DataFrame(results['businesses'])
pb_f2.head() 
```

```python
## concatenate the previous results and new results. 
pb = pd.concat([pb_f, pb_f2],
                      ignore_index=True)
display(pb.head(), pb.tail())
```

```python
# save the results to records-style json file with appropriate name
pb.to_json("Data/yelp-search-plant_based-north_olmsted.json", orient='records')
```
